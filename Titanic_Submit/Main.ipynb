```python
class_age = combined[["Pclass","Age"]]
for i in range(1,4):
    plt.figure(figsize = (10,5))
    plt.title("Class" + str(i))
    y = class_age[class_age["Pclass"] == i]["Age"].value_counts()
    x = y.index.values
    plt.bar(x,y)
    plt.show()

median = class_age.groupby("Pclass").median()
print(median)
```

```python
import numpy as np

names = combined["Name"]

last_name = []
first_name = []
titles = []

for name in names:
    if ',' not in name:
        last_name.append('')
    else:
        last, remaining = name.split(",", 1)
        last_name.append(last.strip())

    if '.' not in remaining:
        titles.append('')
        first_name.append(remaining.strip())
    else:
        title, first = remaining.split('.', 1)
        titles.append(title.strip())
        first_name.append(first.strip())

last_name = np.array(last_name)
first_name = np.array(first_name)
titles = np.array(titles)

idx = np.where(np.isin(titles, ['Capt', 'Col', 'Major']))
titles[idx] = 'Military'
idx = np.where(np.isin(titles, ['Don', 'Dona', 'Jonkheer', 'Lady', 'Sir', 'Master', 'the Countess']))
titles[idx] = 'Nobility'
idx = np.where(np.isin(titles, ['Miss', 'Mlle', 'Ms']))
titles[idx] = 'Ms'
idx = np.where(np.isin(titles, ['Mrs', 'Mme']))
titles[idx] = 'Mrs'

print('\n')
print(np.unique(titles, return_counts=True))

combined['Title'] =  titles
```

```python
bins = np.array([0,10,20,30,40,50,60,70,80])
combined['Age_bin'] = pd.cut(combined['Age'], bins)
```

```python
combined['Num_Family'] = combined['SibSp'] + combined['Parch'] + 1 
```

```python
ticket_dct = dict(combined['Ticket'].value_counts())
combined['tkt_count'] = combined['Ticket'].map(ticket_dct)
display(combined)
```

```python
combined['Fare_per_tkt'] = combined['Fare']/combined['tkt_count']
display(combined)
```

```python
bins = np.array([0,20,40,60,80,150])
combined['Fare_Bin'] = pd.cut(combined['Fare_per_tkt'], bins)
```
